```python
class Car:
    def __init__(self,window,door,enginetype):
        self.windows=window
        self.doors=door
        self.enginetypes=enginetype

    def self_driving(self):
        return "This is a {} car".format(self.enginetypes)
```

```python
car1 = Car(4,5,"petrol")
```

```python
car1.self_driving()
```

```python
car2 = Car(3,4,"diesel")
```

```python
print(car1.windows)
```

## OOPS - Public,Protected and Private

```python
## All the class variables are public
class Car():
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
```

```python
audi=Car(4,5,'Diesel')
```

```python
audi.windows=5
```

```python
audi.windows
```

```python
dir(audi)
```

```python
## All the class variables are protected
class Car():
    def __init__(self,windows,doors,enginetype):
        self._windows=windows
        self._doors=doors
        self._enginetype=enginetype
```

```python
class Truck(Car):
    def __init__(self,windows,doors,enginetype,horsepower):
        super().__init__(windows,doors,enginetype)
        self.horsepower=horsepower
```

```python
truck = Truck(4,4,"Diesel",4000)
dir(truck)
```

```python
truck._doors = 5
```

```python
truck._doors
```

```python
##  protected variables can accessed only from the sub classes
```

```python
audi = Car(4,4,'diesel')
dir(audi)
```

```python
## overridden should only happen from sub class
```

```python
## private
class Car():
    def __init__(self,windows,doors,enginetype):
        self.__windows=windows
        self.__doors=doors
        self.__enginetype=enginetype
```

```python
audi = Car(4,4,"Diesel")
```
