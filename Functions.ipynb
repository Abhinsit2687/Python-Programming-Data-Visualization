```python
## Positional and Keyword Argument
def hello(*args, **kwargs):
    print(args)
    print(kwargs)
```

```python
hello("abhishek","gupta",age=22,dob=2003)
```

```python
lst = ['abhishek','gupta']
dict_args = {'age':29,'dob':1990}
```

```python
hello(*lst,**dict_args)
```

```python
lst = [1,2,3,4,5,6,7]
```

```python
def evenoddsum(lst):
    even_sum = 0
    odd_sum = 0
    for i in lst:
        if i%2 == 0:
            even_sum = even_sum+i
        else:
            odd_sum = odd_sum+i
    return even_sum,odd_sum
```

```python
evenoddsum(lst)
```
