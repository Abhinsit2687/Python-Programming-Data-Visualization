## List Comprehension

List comprehension provide a concise way to create lists. It consists of brackets containing an expression followed bt a for clause, then xero or more for or if clauses. The expressions can be anything, meaning you can put in all kinds of objects in lists.


```python
lst1 = []
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
    
```

```python
lst_square([5,2,3,4,5,6,7])
```

```python
## List Comprehension
```

```python
lst = [1,2,3,4,5,6,7]
```

```python
lst1 = [i*i for i in lst]
```

```python
print(lst1)
```

```python
## List Comprehension
```

```python
lst = [1,2,3,4,5,6,7]
```

```python
lst1 = [i*i for i in lst if i%2 == 0]
```

```python
print(lst1)
```
