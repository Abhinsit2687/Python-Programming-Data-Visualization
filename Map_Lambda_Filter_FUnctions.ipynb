## Map Function in Python

```python
def even_or_odd(num):
    if num%2 == 0:
        return "The number {} is even" .format(num)
    else:
        return "The number {} is odd".format(num)
    
```

```python
even_or_odd(24)
```

```python
lst = [1,2,3,4,5,6,7,8,9,24,56,78]
```

```python
list(map(even_or_odd,lst))
```

## Lambda Function
anonymous function.
A function with no name

```python
def addition(a,b):
    return a+b
```

```python
addition(4,5)
```

```python
addition = lambda a,b: a+b
```

```python
addition(12,14)
```

only have a single expression 

```python
def even(num):
    if num%2 == 0:
        return True
```

```python
even(24)
```

```python
even1 = lambda a : a%2 == 0
```

```python
even1(12)
```

```python
 add = lambda a,b,c : a+b+c
```

```python
add(1,2,3)
```

## Filter Function in Python

```python
def even(num):
    if num%2 == 0:
        return True
```

```python
lst = [1,2,3,4,5,6,7,8,9,0]
```

```python
list(filter(even,lst))
```

```python
list(filter(lambda num:num%2 == 0, lst))
```
